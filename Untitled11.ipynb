```
len(imageList)
```

```

import pandas as pd
gt = pd.read_csv("./myntradataset/styles.csv",error_bad_lines=False)

label_gt = gt[['id', 'gender', 'masterCategory', 'subCategory', 'articleType',
       'baseColour', 'season', 'usage']]
label_gt['id'] = label_gt['id'].astype(str) + '.jpg'

label_gt.to_csv('./myntradataset/labels.csv',index=False)
```

```
label_gt.head()
```

```
corruptImageList = ['39403.jpg','39410.jpg','39401.jpg','39425.jpg','12347.jpg']
```

```

for i in corruptImageList:
    label_gt = label_gt[label_gt.id != i]
```

```
imageList = label_gt.id
```

```

mod_labels = label_gt.set_index('id').T.to_dict('list')
```

```
mod_labels
```

```
with open("./myntradataset/mod_labels.csv",'w') as f:
  f.write("ID,Labels\n")
  for key,value in mod_labels.items():
    value = list(map(str,value))
    labels = " ".join(value)
    f.write("{},{}\n".format(key,labels))
f.close()
```

```
#Set data parameters

ptf.Dataset_Params(dataset_path="./myntradataset/images",
                   path_to_csv="./myntradataset/mod_labels.csv",
                   
                   
                   delimiter = " ",
                   
                   
                   split=0.9,
                   input_size=224, 
                   batch_size=48, 
                   shuffle_data=True,
                   num_processors=3);
```

```
ptf.apply_random_horizontal_flip(train=True, val=True);
ptf.apply_normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], train=True, val=True, test=True);
```

```
ptf.Dataset();
```

```
ptf.system_dict["dataset"]["params"]["classes"]
```

```
ptf.Model_Params(model_name="densenet169", freeze_base_network=False, use_gpu=True, use_pretrained=True);
```

```
ptf.Model();
```

```
ptf.Training_Params(num_epochs=2, display_progress=True, display_progress_realtime=True, 
                    save_intermediate_models=False, save_training_logs=True);
```

```
ptf.optimizer_rmsprop(0.001);
```

```
ptf.lr_fixed();
```

```
ptf.loss_sigmoid_binary_crossentropy()
```

```
ptf.Train();
```
